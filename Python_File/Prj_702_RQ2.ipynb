```python
# === Suppress Warnings ===
import warnings
warnings.filterwarnings('ignore')

# === STEP 1: Import Libraries ===
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid", font_scale=1.1)

# === STEP 2: Manually Define Data from Screenshot ===
data = {
    "restaurant_name": [
        "Janta Snacks", "Domino's Pizza", "Kouzina Kafe - The Food Court", "Happy Brew Cafe",
        "Jaysika DDN Fast Food", "Cafe Yummy", "Huber & Holly", "Krishna Food",
        "ZAATAR SPICE", "Blue Tokai Coffee Roasters"
    ],
    "city": ["Ahmedabad", "Tirupati", "Bangalore", "Tirupati", "Rishikesh", "Raipur", "Ahmedabad", "Ahmedabad", "Bangalore", "Delhi"],
    "total_orders": [1, 1, 2, 1, 1, 1, 1, 1, 1, 1],
    "total_sales": [1510944.0, 1492435.0, 1481783.0, 1477458.0, 1338264.0, 1316921.0, 1283875.0, 1283875.0, 1235347.0, 1228148.0],
    "avg_order_value": [1510944.0, 1492435.0, 740891.5, 1477458.0, 1338264.0, 1316921.0, 1283875.0, 1283875.0, 1235347.0, 1228148.0],
    "avg_rating": [0.0, 3.8, 0.0, 3.5, 3.7, 0.0, 4.5, 0.0, 0.0, 4.7]
}
df = pd.DataFrame(data)

# === STEP 3: Bar Plot - Total Sales by Restaurant ===
plt.figure(figsize=(11, 6))
sns.barplot(data=df.sort_values("total_sales", ascending=False),
            y="restaurant_name", x="total_sales", palette="viridis")
plt.title("Top Restaurants by Total Sales (INR)", fontsize=14)
plt.xlabel("Total Sales")
plt.ylabel("Restaurant")
plt.tight_layout()
plt.show()

print("\nInsight: Janta Snacks and Dominos Pizza top the sales chart. However, nearly all top restaurants show **similar revenue levels**, likely driven by single high-value orders.\n")

# === STEP 4: Bar Plot - Average Order Value ===
plt.figure(figsize=(11, 6))
sns.barplot(data=df.sort_values("avg_order_value", ascending=False),
            y="restaurant_name", x="avg_order_value", palette="crest")
plt.title("Average Order Value per Restaurant (INR)", fontsize=14)
plt.xlabel("Avg Order Value")
plt.ylabel("Restaurant")
plt.tight_layout()
plt.show()

print("\nInsight: Most restaurants here recorded only 1 order. Exception is 'Kouzina Kafe' with 2 orders and lower avg value, suggesting either combo deals or price sensitivity.\n")

# === STEP 5: Dot Plot - Avg Rating vs Total Orders ===
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="total_orders", y="avg_rating",
                hue="restaurant_name", s=100, palette="Set1", legend=False)
plt.title("Average Rating vs Total Orders", fontsize=14)
plt.xlabel("Total Orders")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()

print("\nInsight: Several restaurants have **zero customer ratings**, which could be due to recency, app design, or low user engagement post-order. 'Blue Tokai' and 'Huber & Holly' have strong ratings despite low volume.\n")
```

```python
# === Suppress Warnings ===
import warnings
warnings.filterwarnings('ignore')

# === STEP 1: Import libraries ===
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", font_scale=1.1)
plt.rcParams['figure.figsize'] = (10, 5)

# === STEP 2: Manually define the restaurant data ===
data = {
    "restaurant_name": [
        "MAHARAJA GRILLS & ROLLS",
        "REDDY VISHNU FAST FOODS",
        "SRI LAKSHMI  BAR AND RESTAURANT"
    ],
    "city": ["Bangalore", "Tirupati", "Raichur"],
    "cuisine": ["North Indian", "North Indian", "North Indian,Snacks"],
    "rating": [0.0, 3.6, 2.8],
    "year": [2019, 2018, 2017],
    "total_orders": [1, 1, 1],
    "total_sales": [628657.0, 994954.0, 1089685.0]
}
df = pd.DataFrame(data)

# === STEP 3: Bar Chart - Total Sales per Restaurant ===
plt.figure()
sns.barplot(data=df.sort_values("total_sales", ascending=False),
            x="restaurant_name", y="total_sales", palette="flare")
plt.title("Total Sales for Single-Order Restaurants", fontsize=14)
plt.ylabel("Total Sales (INR)")
plt.xlabel("Restaurant")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

print("\n Insight 1:\nAll three restaurants recorded only one order, yet each had extremely high total salesranging from 6.2L to over 10L.\nThis suggests special events or large group orders.\n")

# === STEP 4: Bar Plot - Customer Rating ===
plt.figure()
sns.barplot(data=df, x="restaurant_name", y="rating", palette="mako")
plt.title("Customer Ratings of Single-Order Restaurants", fontsize=14)
plt.ylabel("Rating")
plt.xlabel("Restaurant")
plt.xticks(rotation=15)
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

print("\n Insight 2:\n'MAHARAJA GRILLS & ROLLS' had no rating despite high-value sales.\nThe best-rated among these is 'REDDY VISHNU FAST FOODS' with a decent 3.6 rating.\n")

# === STEP 5: Line Plot - Sales vs Order Year ===
plt.figure()
sns.lineplot(data=df.sort_values("year"), x="year", y="total_sales", marker="o", linewidth=2.5)

plt.title("Year-wise Total Sales (Single Orders)", fontsize=14)
plt.ylabel("Total Sales")
plt.xlabel("Order Year")
plt.grid(True)

# Fix the x-axis to show only full years
plt.xticks(df["year"].sort_values().unique().astype(int))

plt.tight_layout()
plt.show()

print("\n Insight 3:\nThere seems to be no clear time trendhigh-value orders were placed in 2017, 2018, and 2019.\nThis supports the assumption that such orders are isolated events rather than repeat purchases.\n")
```
